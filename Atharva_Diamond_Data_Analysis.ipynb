```python
# So we will consider only that columns
```

```python
column_num = ['carat','depth', 'table', 'x', 'y', 'z','price' ]
for i in column_num:
    sns.boxplot(df[i],whis =1.5)
    plt.show();
```

```python
# so as we can observe there are few outliers occuring in our numerical columns so to remove those we can use the standard
# function.
```

```python
def remove_outlier(colo):
    sorted(colo)
    Q1,Q3=np.percentile(colo,[25,75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range
```

```python
for column in df[column_num].columns:
    lr,ur=remove_outlier(df[column])
    df[column]=np.where(df[column]>ur,ur,df[column])
    df[column]=np.where(df[column]<lr,lr,df[column])
```
